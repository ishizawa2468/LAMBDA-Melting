```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import LogLocator

from modules.file_format.HDF5 import HDF5Reader
from modules.file_format.spe_wrapper import SpeWrapper

plt.rcParams['font.size'] = 14
path = '/Users/ishizawaosamu/work/MasterThesis/data/BDD/OIbDia08/2nd/OIbDia08_ 04_separate_half_p45e-2_dist.hdf'
spe_path = '/Users/ishizawaosamu/work/MasterThesis/data/BDD/OIbDia08/2nd/OIbDia08_ 04.spe'
```

```python
dist_hdf = HDF5Reader(path)
```

```python
# 温度map, エラーjとradiation強度を読み込む
dist_array = dist_hdf.find_by(query='value/T')
dist_array[dist_array > 10_000] = 0 # ここで変な値は切っておく

T_fitting_error = dist_hdf.find_by(query='error/T')

radiation_array = dist_hdf.find_by(query='spe/2d_max_intensity')
```

```python
spe = SpeWrapper(spe_path)
spe.get_params_from_xml()
spe.framerate
```

```python
# 温度データの形状を取得しておく
frame_num, position_num = dist_array.shape
# 配列を作っておく
frame_array = np.arange(frame_num)
position_array = np.arange(position_num)
frame_num, position_num
```

```python
frame = 350

fig, main_ax = plt.subplots(figsize=(6, 8), dpi=300)

# 温度データ (左の軸)
main_ax.scatter(position_array, dist_array[frame], color='red', s=10, label='Temperature')
main_ax.plot(position_array, dist_array[frame], color='red', alpha=0.5)
main_ax.set_xlabel('Position')
main_ax.set_ylabel('Temperature', color='red')
main_ax.tick_params(axis='y', labelcolor='red')
main_ax.set_xlim(360, 410)

# 放射データ (右の軸)
twin = main_ax.twinx()
twin.scatter(position_array, radiation_array[frame], color='royalblue', s=10, label='Radiation')
twin.plot(position_array, radiation_array[frame], color='royalblue', alpha=0.5)
twin.set_ylabel('Radiation', color='royalblue')
twin.tick_params(axis='y', labelcolor='royalblue')

# 凡例を結合して表示
handles_main, labels_main = main_ax.get_legend_handles_labels()
handles_twin, labels_twin = twin.get_legend_handles_labels()
main_ax.legend(handles_main + handles_twin, labels_main + labels_twin, loc='upper right')

plt.show()
plt.close()
```

```python
frames = range(350, 750)

position_array_length = position_array * 1.28

for frame in frames:
    fig, main_ax = plt.subplots(figsize=(10, 8), dpi=300)

    # 温度データ (左の軸)
    main_ax.scatter(position_array_length, dist_array[frame], color='red', s=10, label='Temperature')
    main_ax.plot(position_array_length, dist_array[frame], color='red', alpha=0.5)
    main_ax.set_xlabel('Position (μm)')
    main_ax.set_ylabel('Temperature (K)', color='red')
    main_ax.tick_params(axis='y', labelcolor='red')
    main_ax.set_xlim(360*1.28, 410*1.28)
    main_ax.set_ylim(2_000, 7_000)

    # 放射データ (右の軸)
    twin = main_ax.twinx()
    twin.scatter(position_array_length, radiation_array[frame], color='royalblue', s=10, label='Radiation')
    twin.plot(position_array_length, radiation_array[frame], color='royalblue', alpha=0.5)
    twin.set_ylabel('Radiation (a.u.)', color='royalblue')
    twin.tick_params(axis='y', labelcolor='royalblue')

    # Set log scale for the right y-axis
    twin.set_yscale('log')
    twin.set_ylim(500, 100_000)
    twin.yaxis.set_major_locator(LogLocator(base=10.0, subs=None, numticks=10))  # Log scale major ticks

    # 凡例を結合して表示
    handles_main, labels_main = main_ax.get_legend_handles_labels()
    handles_twin, labels_twin = twin.get_legend_handles_labels()
    main_ax.legend(handles_main + handles_twin, labels_main + labels_twin, loc='upper right')
    main_ax.set_title(f'Temperature vs Radiation / Time = {frame/spe.framerate:.3f} seconds')
    plt.savefig(f'/Users/ishizawaosamu/work/MasterThesis/save/fig/OI08_2nd_T_radi/T_distribution_{frame}.png', dpi=300)

    plt.show()
    plt.close()
```
